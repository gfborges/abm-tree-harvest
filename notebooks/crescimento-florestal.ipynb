```python
pos_cuts = pd.read_csv("../dados/output/cuts-oliveira-gouveia.csv")
neg_cuts = pd.read_csv("../dados/output/cuts-borges-gouveia.csv")
```

```python
pos_cuts.head()
```

```python
neg_cuts.head()
```

```python
comp=  pd.DataFrame()

comp["media"] =  neg_cuts["cut"].astype("int")[:-1]
comp["media_vol"] =  neg_cuts["cut_vol"][:-1].astype("float")
comp["Neg_vol_tree"] = comp["media_vol"].values /722
comp["Positivo"] =  pos_cuts["cut"].astype("int")[:-1]
comp["Positivo_vol"] =  pos_cuts["cut_vol"][:-1].astype("float")
comp["Pos_vol_tree"] = comp["Positivo_vol"] /722
comp.head()
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
ax1.bar(range(2, 4), height = comp["media"][1:])
ax2.bar(range(2, 4), height = comp["Positivo"][1:])
plt.suptitle("Comparação dos cortes", fontdict ={"fontsize":18, "fontweight":"bold"})
ax1.set_title("media")
ax2.set_title("Positivo")
ax1.set_xticks([2, 3])
ax2.set_xticks([2, 3])
ax1.set_xlabel("Cortes")
ax2.set_xlabel("Cortes")
plt.savefig("cut-comp.png", dpi= 250)
plt.show()
```
