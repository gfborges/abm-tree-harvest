```python
dead = pd.DataFrame({
    "Mortalidade": [4.7,4.18,8   ,6.7  ,11.64,21.3,  9],
    "Recrutamento":[5.1,3.68,3.49,11.35,11.87,26.37, 7],
}, index= ["Higuchi(2004)", "MPEG(2004)", "Colpini(2010)","Souza et al(2012)", "Andrade(2018)0", "Andrade (2018)1", "Santos(2018)"])
```

```python
dead.head()
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.barplot(x=dead.index ,y="Mortalidade", data=dead)
```

```python
dead.reset_index(inplace=True)
dead.rename(columns={'index':'author'}, inplace=True)
```

```python
deadR = dead[['author', 'Recrutamento']]
deadR.columns = ['author', 'Value']
deadR["Tipo"] = ["Recrutamento"]*7
deadM = dead[['author', 'Mortalidade']]
deadM.columns = ['author', 'Value']
deadM["Tipo"] = ["Mortalidade"]*7
display(deadR)
```

```python
dead2 = pd.concat([deadR, deadM])
dead2
```

```python
sns.set()
sns.barplot(y='Value', x='author', hue='Tipo', data= dead2)
plt.title("Mortalidade e Recrutamento", fontdict={"fontsize":14})
plt.savefig("morte-recrutamento.png", dpi= 300)
plt.show()
```

```python
def make_final_chart(df, name, year=9):
    df = df.loc[year,"class1":"class9"]
    sns.set()
    sns.barplot(x=df.index,y=df.values)
    plt.title("Quantidade de árvore em cada classe", fontdict ={"fontsize":14})
    plt.savefig(name, dpi=300)
    plt.show()
```

```python
arvs = bg_prag.loc[3,'class1':'class9']
arvs
```

```python
area_arv = sum([np.pi*(d/2)**2 for d in arvs ])
```

```python
area_arv
```

```python
def area_arv(df, year=9):
    arvs = df.loc[year,'class1':'class9']
    return sum([np.pi*d*(((c+1)*10-5)/2)**2  for c,d in enumerate(arvs)])
```

```python
print("area_arv")
print("oliviera:",area_arv(ol_prag,3))
print("borges:",area_arv(bg_prag,3))
print("proporção:", area_arv(ol_prag,3)  / area_arv(bg_prag,3))
```
